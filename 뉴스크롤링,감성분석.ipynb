```python
# 라이브러리 불러오기
from selenium import webdriver # 브라우저 자동화를 위한 모듈
import requests # HTTP 요청을 보내기 위한 모듈
from bs4 import BeautifulSoup as BS # HTML 내용 파싱을 위한 모듈
import pandas as pd # 데이터 조작 및 분석을 모듈
import time # 코드 실행 속도 조절을 위한 모듈
import re # 정규 표현식 사용을 위한 모듈
import pickle # 파이썬 객체 직렬화를 위한 모듈
from selenium.webdriver.common.by import By # 다양한 방법으로 엘리먼트를 찾기 위한 모듈
import datetime # 날짜와 시간 연산을 위한 모듈
from newspaper import Article

# 파일 저장 경로 설정
file_path = 'C:\\Users\\alsrb\\'

# 웹 드라이버 설정 및 네이버 뉴스 페이지로 이동
driver = webdriver.Chrome()
driver.get('https://search.naver.com/search.naver?where=news&query=%EB%8D%94%EB%B6%88%EC%96%B4%EB%AF%BC%EC%A3%BC%EB%8B%B9&sm=tab_opt&sort=0&photo=0&field=0&pd=3&ds=2020.09.02&de=2021.01.03&docid=&related=0&mynews=1&office_type=1&office_section_code=1&news_office_checked=1056&nso=so%3Ar%2Cp%3Afrom20200902to20210103&is_sug_officeid=0&office_category=0&service_area=0')

# 스크롤 다운 함수 정의
def doScrollDown(whileSeconds):
    start = datetime.datetime.now() # 스크롤 다운 시작 시간 설정
    end = start + datetime.timedelta(seconds=whileSeconds) # 스크롤 다운 종료 시간 설정
    while True:
        # 페이지 맨 아래로 스크롤 다운
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(0.3) 
        
        # 종료 시간에 도달하면 반복 종료
        if datetime.datetime.now() > end:  
            break

# 페이지 스크롤 다운
doScrollDown(120)

# 제목과 URL을 저장할 리스트 초기화
title_list = []  # 제목을 저장할 리스트 초기화
url_list = []  # URL을 저장할 리스트 초기화

# 현재 페이지에서 class가 'news_tit'인 요소를 찾아 제목과 URL 추출
titles = driver.find_elements(By.CLASS_NAME, 'news_tit')
for i, title_element in enumerate(titles):
    if i == 1000:  # 1000개까지만 수집
        break
    try:
        title_list.append(title_element.text)  
        url_list.append(title_element.get_attribute('href'))  
    except:
        print("오류 발생")
        continue  

    if (i+1) % 10 == 0: 
        print(f'{i+1}개 뉴스 제목과 URL을 수집했습니다.')

# 뉴스 본문 저장할 리스트 초기화
content_list = []

# 수집한 URL에서 뉴스 본문 추출
print("뉴스 본문 추출을 시작합니다.")
for i, url in enumerate(url_list):
    try:
        article = Article(url, language='ko')
        article.download()
        article.parse()
        content_list.append(article.text)
    except:
        print(f"오류 발생 - URL: {url}")
        content_list.append('')  
        continue
    
    print(f'{i+1}/{len(url_list)} 뉴스 본문을 추출했습니다.')

print("뉴스 본문 추출이 완료되었습니다.")

# 데이터프레임 생성
news_df = pd.DataFrame({
    'Title': title_list,
    'Text': content_list,  # 'Text'로 변경
    'Query': '더불어민주당',  # 'Query' 컬럼 추가
    'News': 'kbs'  # 'News' 컬럼 추가
})

# 데이터프레임을 CSV 파일로 저장
csv_file_path = file_path + 'DemocraticParty_kbs_01.csv'
news_df.to_csv(csv_file_path, index=False, encoding='utf-8-sig')

print(f'뉴스 데이터를 {csv_file_path}에 저장했습니다.')

# 데이터프레임을 pickle 파일로 저장
pickle_file_path = file_path + 'DemocraticParty_kbs_01.pkl'
with open(pickle_file_path, 'wb') as file:
    pickle.dump(news_df, file)
    
print(f'뉴스 데이터를 {pickle_file_path}에 저장했습니다.')

# 웹 드라이버 종료
driver.quit()
```

```python
# 라이브러리 불러오기
from selenium import webdriver # 브라우저 자동화를 위한 모듈
import requests # HTTP 요청을 보내기 위한 모듈
from bs4 import BeautifulSoup as BS # HTML 내용 파싱을 위한 모듈
import pandas as pd # 데이터 조작 및 분석을 모듈
import time # 코드 실행 속도 조절을 위한 모듈
import re # 정규 표현식 사용을 위한 모듈
import pickle # 파이썬 객체 직렬화를 위한 모듈
from selenium.webdriver.common.by import By # 다양한 방법으로 엘리먼트를 찾기 위한 모듈
import datetime # 날짜와 시간 연산을 위한 모듈
from newspaper import Article

# 파일 저장 경로 설정


driver = webdriver.Chrome()
driver.get('https://search.naver.com/search.naver?where=news&query=%EB%8D%94%EB%B6%88%EC%96%B4%EB%AF%BC%EC%A3%BC%EB%8B%B9&sm=tab_opt&sort=0&photo=0&field=0&pd=3&ds=2020.09.02&de=2020.12.01&docid=&related=0&mynews=1&office_type=1&office_section_code=3&news_office_checked=1015&nso=so%3Ar%2Cp%3Afrom20200902to20201201&is_sug_officeid=0&office_category=0&service_area=0')

# 스크롤 다운 함수 정의
def doScrollDown(whileSeconds):
    start = datetime.datetime.now() # 스크롤 다운 시작 시간 설정
    end = start + datetime.timedelta(seconds=whileSeconds) # 스크롤 다운 종료 시간 설정
    while True:
        # 페이지 맨 아래로 스크롤 다운
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(0.3) # 1초 대기
        
        # 종료 시간에 도달하면 반복 종료
        if datetime.datetime.now() > end:  
            break

# 페이지 스크롤 다운
doScrollDown(120)

# 제목과 URL을 저장할 리스트 초기화
title_list = []  # 제목을 저장할 리스트 초기화
url_list = []  # URL을 저장할 리스트 초기화

# 현재 페이지에서 class가 'news_tit'인 요소를 찾아 제목과 URL 추출
titles = driver.find_elements(By.CLASS_NAME, 'news_tit')
for i, title_element in enumerate(titles):
    if i == 1000:  # 1000개까지만 수집
        break
    try:
        title_list.append(title_element.text)  
        url_list.append(title_element.get_attribute('href'))  
    except:
        print("오류 발생")
        continue  

    if (i+1) % 10 == 0: 
        print(f'{i+1}개 뉴스 제목과 URL을 수집했습니다.')

# 뉴스 본문 저장할 리스트 초기화
content_list = []

# 수집한 URL에서 뉴스 본문 추출
print("뉴스 본문 추출을 시작합니다.")
for i, url in enumerate(url_list):
    try_count = 0
    while try_count < 2:
        try:
            article = Article(url, language='ko')
            article.download()
            time.sleep(1)  # 다운로드 후 대기 시간 추가
            article.parse()
            content_list.append(article.text)
            break
        except:
            try_count += 1
            print(f"오류 발생 - URL: {url}, 재시도 중입니다 ({try_count}/2)")
            if try_count == 2:
                print(f"최종 오류 발생 - URL: {url}, 다음으로 넘어갑니다.")
                content_list.append('')  
    
    print(f'{i+1}/{len(url_list)} 뉴스 본문을 추출했습니다.')

print("뉴스 본문 추출이 완료되었습니다.")

# 데이터프레임 생성
news_df = pd.DataFrame({
    'Title': title_list,
    'Text': content_list,  # 'Text'로 변경
    'Query': '더불어민주당',  # 'Query' 컬럼 추가
    'News': '한국경제'  # 'News' 컬럼 추가
})
# 데이터프레임을 CSV 파일로 저장
csv_file_path = file_path + 'DemocraticParty_hankyung_01.csv'
news_df.to_csv(csv_file_path, index=False, encoding='utf-8-sig')

print(f'뉴스 데이터를 {csv_file_path}에 저장했습니다.')

# 데이터프레임을 pickle 파일로 저장
pickle_file_path = file_path + 'DemocraticParty_hankyung_01.pkl'
with open(pickle_file_path, 'wb') as file:
    pickle.dump(news_df, file)
    
print(f'뉴스 데이터를 {pickle_file_path}에 저장했습니다.')

# 웹 드라이버 종료
driver.quit()
```

```python
# 라이브러리 불러오기
from selenium import webdriver # 브라우저 자동화를 위한 모듈
import requests # HTTP 요청을 보내기 위한 모듈
from bs4 import BeautifulSoup as BS # HTML 내용 파싱을 위한 모듈
import pandas as pd # 데이터 조작 및 분석을 모듈
import time # 코드 실행 속도 조절을 위한 모듈
import re # 정규 표현식 사용을 위한 모듈
import pickle # 파이썬 객체 직렬화를 위한 모듈
from selenium.webdriver.common.by import By # 다양한 방법으로 엘리먼트를 찾기 위한 모듈
import datetime # 날짜와 시간 연산을 위한 모듈
from newspaper import Article



# 웹 드라이버 설정 및 네이버 뉴스 페이지로 이동
driver = webdriver.Chrome()
driver.get('https://search.naver.com/search.naver?where=news&query=%EA%B5%AD%EB%AF%BC%EC%9D%98%ED%9E%98&sm=tab_opt&sort=0&photo=0&field=0&pd=3&ds=2020.09.02&de=2020.12.01&docid=&related=0&mynews=1&office_type=1&office_section_code=2&news_office_checked=1056&nso=so%3Ar%2Cp%3Afrom20200902to20201201&is_sug_officeid=0&office_category=0&service_area=0')

# 스크롤 다운 함수 정의
def doScrollDown(whileSeconds):
    start = datetime.datetime.now() # 스크롤 다운 시작 시간 설정
    end = start + datetime.timedelta(seconds=whileSeconds) # 스크롤 다운 종료 시간 설정
    while True:
        # 페이지 맨 아래로 스크롤 다운
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(0.3) 
        
        # 종료 시간에 도달하면 반복 종료
        if datetime.datetime.now() > end:  
            break

# 페이지 스크롤 다운
doScrollDown(250)

# 제목과 URL을 저장할 리스트 초기화
title_list = []  # 제목을 저장할 리스트 초기화
url_list = []  # URL을 저장할 리스트 초기화

# 현재 페이지에서 class가 'news_tit'인 요소를 찾아 제목과 URL 추출
titles = driver.find_elements(By.CLASS_NAME, 'news_tit')
for i, title_element in enumerate(titles):
    if i == 1000:  # 1000개까지만 수집
        break
    try:
        title_list.append(title_element.text)  
        url_list.append(title_element.get_attribute('href'))  
    except:
        print("오류 발생")
        continue  

    if (i+1) % 10 == 0: 
        print(f'{i+1}개 뉴스 제목과 URL을 수집했습니다.')

# 뉴스 본문 저장할 리스트 초기화
content_list = []

# 수집한 URL에서 뉴스 본문 추출
print("뉴스 본문 추출을 시작합니다.")
for i, url in enumerate(url_list):
    try:
        article = Article(url, language='ko')
        article.download()
        article.parse()
        content_list.append(article.text)
    except:
        print(f"오류 발생 - URL: {url}")
        content_list.append('')  
        continue
    
    print(f'{i+1}/{len(url_list)} 뉴스 본문을 추출했습니다.')

print("뉴스 본문 추출이 완료되었습니다.")

# 데이터프레임 생성
news_df = pd.DataFrame({
    'Title': title_list,
    'Text': content_list,  # 'Text'로 변경
    'Query': '국민의힘',  # 'Query' 컬럼 추가
    'News': 'kbs'  # 'News' 컬럼 추가
})

# 데이터프레임을 CSV 파일로 저장
csv_file_path = file_path + 'PeoplePowerParty_kbs_01.csv'
news_df.to_csv(csv_file_path, index=False, encoding='utf-8-sig')

print(f'뉴스 데이터를 {csv_file_path}에 저장했습니다.')

# 데이터프레임을 pickle 파일로 저장
pickle_file_path = file_path + 'PeoplePowerParty_kbs_01.pkl'
with open(pickle_file_path, 'wb') as file:
    pickle.dump(news_df, file)
    
print(f'뉴스 데이터를 {pickle_file_path}에 저장했습니다.')

# 웹 드라이버 종료
driver.quit()
```

```python
s_word = []  
values = []
score = []

def average(list):  # 각 문서의 평균 감성 score 반환  # 리스트의 평균 값을 계산하는 함수
    return sum(list)/len(list)

# Text와 Title 결합 후 감성 분석 수행
for index, row in tqdm(data.iterrows(), total=len(data), desc="Processing documents"):
    combined_text = ' '.join(row['Text']) + ' ' + ' '.join(row['Title'])  # Text와 Title 결합
    temp_s_word = []  # 문서에서 발견된 감성 단어 리스트 초기화
    temp_value = []

    
# 감성사전에 있는 단어와 현재 문서의 단어 비교
    for s in sentiword:
        if s['word'] in combined_text:
            if len(s['word']) > 1 : # 한글자 이상의 감성단어와 비교 
                temp_s_word.append(s['word'])
                temp_value.append(int(s['polarity'])) 
    s_word.append(temp_s_word) # 해당하는 감성단어 추가
    values.append(temp_value) # 해당 단어의 polarity값 추가

    try:
        score.append(average(temp_value)) # 각 문서의 평균 polarity 추가
    except ZeroDivisionError:
        score.append(int(0))

# 결과 데이터프레임에 감성분석 결과 삽입
data=data.assign(sentiword = s_word,values=values,score=score) # 결과 데이터프레임에 삽입
```

```python
import pandas as pd
import numpy as np

# 데이터 로드

data = pd.read_feather(file_path + "emotion.feather")


# 각 문서별 감성 분석 결과를 저장할 컬럼 초기화
data['positive_count'] = 0
data['negative_count'] = 0
data['positive_ratio'] = 0.0
data['negative_ratio'] = 0.0

# 각 문서별 감성 분석 수행
for index, row in tqdm(data.iterrows(), total=len(data), desc="Processing documents"):
    positive_count = sum(np.array(row['values']) > 0)
    negative_count = sum(np.array(row['values']) < 0)
    
    total_count = len(row['values'])
    
    if total_count > 0:
        data.at[index, 'positive_count'] = positive_count
        data.at[index, 'negative_count'] = negative_count
        data.at[index, 'positive_ratio'] = positive_count / total_count
        data.at[index, 'negative_ratio'] = negative_count / total_count

# 결과 저장
data.to_feather(file_path + "emotion_data.feather")

# 데이터 확인
data.head()
```

```python
# 새로운 열 추가: 긍정, 중립, 부정 카테고리
data['sentiment_category'] = ''

# 감성 카테고리 결정 함수
def categorize_sentiment(score):
    if score > 0:
        return '긍정'
    elif score < 0:
        return '부정'
    else:
        return '중립'

# 각 문서별로 감성 카테고리 적용
data['sentiment_category'] = data['score'].apply(categorize_sentiment)

# 결과 저장
data.to_feather(file_path + "emotion_data2.feather")

# 데이터 확인
data
```
